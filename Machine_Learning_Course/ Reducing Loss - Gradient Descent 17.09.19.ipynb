```python
def iterative_loss_reduction(x,y):
    test_w = 0
    test_b = 0
    lobf = w*x.T+b
    test_lobf = test_w*x.T+b
    plt.plot(x, y.T, 'ro')
    plt.plot(x.T, lobf.T, 'b')
    plt.plot(x.T, test_lobf.T, 'g')
    #calculate MSE for blue line
    squared_loss = [i*i for i in (lobf-y)]
    MSE = (np.sum(squared_loss))/(len(squared_loss[0]))
    print(MSE)
    #calculate MSE for green line
    squared_loss_TEST = [i*i for i in (test_lobf-y)]
    MSE_TEST = (np.sum(squared_loss_TEST))/(len(squared_loss_TEST[0]))
    print(MSE_TEST)
    MSE_ideal = 2
    while MSE_TEST > MSE_ideal:
        test_w += 0.05
        test_lobf = test_w*x.T+b
        plt.plot(x, y.T, 'ro')
        plt.plot(x.T, lobf.T, 'b')
        plt.plot(x.T, test_lobf.T, 'g')
        squared_loss_TEST = [i*i for i in (test_lobf-y)]
        MSE_TEST = (np.sum(squared_loss_TEST))/(len(squared_loss_TEST[0]))
        print(MSE_TEST)
   
```

instead of above, i want to iterate over values of w, to find a local minimum MSE.

```python
local_minimum = []
for num in np.arange(0,2.05,0.05):
    test_w = num
    test_lobf = test_w*x.T+b
    plt.plot(x, y.T, 'ro')   
    plt.plot(x.T, lobf.T, 'b')
    plt.plot(x.T, test_lobf.T, 'g')
    squared_loss_TEST = [i*i for i in (test_lobf-y)]
    MSE_TEST = (np.sum(squared_loss_TEST))/(len(squared_loss_TEST[0]))
    print(MSE_TEST)
    local_minimum.append([test_w, MSE_TEST])
```

```python
for x,y in local_minimum:
    plt.plot(x,y, 'ro')
```
