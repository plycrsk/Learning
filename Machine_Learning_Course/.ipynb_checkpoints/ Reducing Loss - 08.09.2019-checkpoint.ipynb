```python
from matplotlib import pyplot as plt
import numpy as np
import math

x = np.arange(0,20) #generate array x with numbers 0 to 19
x2 = np.array([x, np.ones(20)]) #generate 2d array x2 with 0 to 19 in 1d, and 1's in 2d
y = x + np.random.random((1,20))
w,b = np.linalg.lstsq(x2.T,y.T)[0]
```

```python
lobf = w*x.T+b
plt.plot(x, y.T, 'ro')
plt.plot(x.T, lobf.T, 'b')
```

```python
squared_loss = [i*i for i in (lobf-y)]
MSE = (np.sum(squared_loss))/(len(squared_loss[0]))
print(MSE)
```

```python
#now i will test effect of w on MSE
#generate random list of w's between 0 and 1
test_ws = np.random.random(20)
```

```python
test_ws
```

```python
for i in test_ws:
    lobf_test = i*x.T+b
    #print(lobf_test)
    squared_loss_test = [z*z for z in (lobf_test-y)]
    #print(squared_loss_test)
    MSE_test = (np.sum(squared_loss_test))/(len(squared_loss_test[0]))
    print(MSE_test)
```

```python
#now i will try to make a function to reduce loss iteratively
def iterative_loss_reduction(x,y):
    x = np.arange(0,20) #generate array x with numbers 0 to 19
    x2 = np.array([x, np.ones(20)]) #generate 2d array x2 with 0 to 19 in 1d, and 1's in 2d
    y = x + 5*np.random.random((1,20))-1
    w,b = np.linalg.lstsq(x2.T,y.T)[0]
    test_w = 0
    test_b = 0
    lobf = w*x.T+b
    plt.plot(x, y.T, 'ro')
    plt.plot(x.T, lobf.T, 'b')
    plt.plot(x.T, test_w*x.T+test_b, 'g')
    return w
```

```python
iterative_loss_reduction(x,y)
```

above.. I will now train the program to find the best fit line (blue), from starting point green.

```python
test_w = 0
test_b = 0
x = np.arange(0,20) #generate array x with numbers 0 to 19
x2 = np.array([x, np.ones(20)]) #generate 2d array x2 with 0 to 19 in 1d, and 1's in 2d
y = x + 5*np.random.random((1,20))-1
```

```python
def iterative_loss_reduction(x,y):
    test_w = 0
    test_b = 0
    lobf = w*x.T+b
    test_lobf = test_w*x.T+b
    plt.plot(x, y.T, 'ro')
    plt.plot(x.T, lobf.T, 'b')
    plt.plot(x.T, test_lobf.T, 'g')
    #calculate MSE for blue line
    squared_loss = [i*i for i in (lobf-y)]
    MSE = (np.sum(squared_loss))/(len(squared_loss[0]))
    print(MSE)
    #calculate MSE for green line
    squared_loss_TEST = [i*i for i in (test_lobf-y)]
    MSE_TEST = (np.sum(squared_loss_TEST))/(len(squared_loss_TEST[0]))
    print(MSE_TEST)
    MSE_ideal = 2
    while MSE_TEST > MSE_ideal:
        test_w += 0.05
        test_lobf = test_w*x.T+b
        plt.plot(x, y.T, 'ro')
        plt.plot(x.T, lobf.T, 'b')
        plt.plot(x.T, test_lobf.T, 'g')
        squared_loss_TEST = [i*i for i in (test_lobf-y)]
        MSE_TEST = (np.sum(squared_loss_TEST))/(len(squared_loss_TEST[0]))
        print(MSE_TEST)
   
```

```python
iterative_loss_reduction(x,y)
```
